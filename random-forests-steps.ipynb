```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python
hit = pd.read_csv("../input/hittlers/Hitters.csv")
df = hit.copy()
df = df.dropna()
dms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

```python
# MODEL
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model= RandomForestRegressor(random_state=42)
```

```python
rf_model.fit(X_train,y_train)
```

# Parameters
### n_estimators: the number of trees whic will be predicted***
### max_features: max features which are considered in every division***
### max_depth:  avoid overfitting ** 
### max_features: max feature number which is considered in division operations
### min_sample_split: min number of observations in the node before this node is divided
### min_samples_leaf: min number of observations in leaf node

```python
#Prediction
```

```python
y_pred=rf_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_pred, y_test))
```

```python
#Tuning
```

```python
rf_params = {'max_depth': list(range(1,10)),
            'max_features': [3,5,10,15],
            'n_estimators' : [100, 200, 500, 1000, 2000]}
```

```python
rf_model = RandomForestRegressor(random_state = 42)
```

```python
rf_cv_model = GridSearchCV(rf_model, 
                           rf_params, 
                           cv = 10, 
                            n_jobs = -1)  #it provides full performance to operator
```

```python
rf_cv_model.fit(X_train, y_train)
```

```python
rf_cv_model.best_params_
```

```python
rf_tuned = RandomForestRegressor(max_depth  = 8, 
                                 max_features = 3, 
                                 n_estimators =200)
```

```python
rf_tuned.fit(X_train, y_train)
```

```python
y_pred = rf_tuned.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```
